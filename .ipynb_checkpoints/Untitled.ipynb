```python
# Training and Testing Data Split
data_training = pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
data_testing = pd.DataFrame(df['Close'][int(len(df)*0.70):])
```

```python
data_training.shape
```

```python
data_testing.shape
```

```python
# Normalize Data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
data_training_array = scaler.fit_transform(data_training)
```

```python
x_train, y_train = [], []

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])  
    y_train.append(data_training_array[i, 0])  # Fix here: y_train should take single value, not a slice

x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
#Model building 
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
model = Sequential()

model.add(LSTM(units=50, activation='tanh', return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(units=60, activation='tanh', return_sequences=True))
model.add(Dropout(0.3))

model.add(LSTM(units=80, activation='tanh', return_sequences=True))
model.add(Dropout(0.4))

model.add(LSTM(units=120, activation='tanh'))
model.add(Dropout(0.5))

model.add(Dense(units=1))
```

```python
model.summary()
```

```python
# Compile Model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train Model
model.fit(x_train, y_train, epochs=50)
```

```python
past_100_days = data_training.tail(100)
```

```python
final_df = pd.concat([past_100_days, data_testing], ignore_index=True)  
```

```python
final_df.head()
```

```python
input_data = scaler.fit_transform(final_df)
```

```python
x_test, y_test = [], []

for i in range(100, input_data.shape[0]):
    x_test.append(input_data[i-100:i])  
    y_test.append(input_data[i, 0])  # Fix here: y_train should take single value, not a slice

x_test, y_test = np.array(x_test), np.array(y_test)
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted.shape
```

```python
scaler.scale_
```

```python
scaler_factor = 1/ 0.00339719
y_predicted = y_predicted * scaler_factor
y_test = y_test * scaler_factor
```

```python
plt.figure(figsize=(12,4))
plt.plot(y_test, label = 'original price', linewidth = 1)
plt.plot(y_predicted, label = 'predicted price', linewidth =1)
plt.legend()
plt.show()
```
